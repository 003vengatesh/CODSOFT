```python
import numpy as np
import pandas as pd
import nltk
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from sklearn.feature_extraction.text import CountVectorizer
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import BernoulliNB
from sklearn.metrics import classification_report
```

```python
corpus = []

for i in range(len(data['descript'])):
    review = re.sub('[^a-zA-Z0-9]', ' ', data.descript[i])
    review = review.lower()
    corpus.append(review)
```

```python
# Lemmatization
nltk.download('punkt')
nltk.download('stopwords')

lemmatizer = WordNetLemmatizer()
nltk.download('wordnet')
nltk.download('omw-1.4')
for i in corpus:
    words = nltk.word_tokenize(i)
    for word in words:
        if word not in set(stopwords.words('english')):
            print(lemmatizer.lemmatize(word))
```

```python
# Removing Stopwords

stop_words = set(stopwords.words('english'))

corpus = data['descript']
def remove_stopwords(text):
    tokens = text.split()
    filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
    return ' '.join(filtered_tokens)

corpus = corpus.apply(remove_stopwords)
data['descript'] = corpus
```

```python
corpus = data['descript']
tfidf_vectorizer = TfidfVectorizer(max_features = 2500, ngram_range = (1,2))
tfidf_matrix = tfidf_vectorizer.fit_transform(corpus)
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns = tfidf_vectorizer.get_feature_names_out())

corpus = pd.concat([corpus, tfidf_df], axis = 1)
```

```python
corpus.drop(columns = 'descript', axis = 1, inplace = True)
```

```python
data = pd.concat([data, corpus], axis = 1)
```

```python
data = pd.read_csv('total_preprocessed.csv')
```

```python
# Train Test Split

x = data.drop(columns = 'genre', axis = 1)
y = data['genre']

x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.3, random_state = 1)
```

MODEL SELECTION

Naive Bayes

```python
#Gaussian naive-bayes

model = GaussianNB()

model.fit(x_train, y_train)
y_pred = model.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
# Multinomial naive-bayes

mul_model = MultinomialNB()
mul_model.fit(x_train, y_train)
y_pred = mul_model.predict(x_test)
accuracy_score(y_pred, y_test)
```

```python
# Bernaulli naive-bayes

ber_model= BernoulliNB()
ber_model.fit(x_train, y_train)
y_pred = ber_model.predict(x_test)
accuracy_score(y_pred, y_test)
```

Logistic Regression

MODEL

```python
# LOGISTIC REGRESSION
lr_model = LogisticRegression(max_iter = 300)
lr_model.fit(x_train,y_train)

lr_model.score(x_test, y_test)
y_pred = lr_model.predict(x_test)
classification_report(y_test, y_pred)
```
